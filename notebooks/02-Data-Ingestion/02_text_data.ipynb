# **Importación de texto en Python**

La función `open(<filename>, <mode>)` permite abrir el archivo llamado <filename>; <mode> es una cadena de texto que especifica la forma en que se habre el archivo asi: `r` para lectura, `w` para escritura (borra el contenido del archivo) y `a` para escribir al final (append); adicionalmente, se puede agregar `b` para indicar que el archivo se abre en modo binario (por defecto se abre en modo texto). `open` devuelve un objeto que permite manipuar el archivo.



```python
#
# Lectura del archivo con open()
# ===============================================
# La función readline() lee una linea a la vez
#
filename = "raw/shakespeare.txt"
file = open(filename, mode="r")
extract = file.readline()
```

```python
print(extract)
```

```python
#
# Se lee el resto del archivo con read()
# ===============================================
#
text = file.read()
print(text)
```

```python
#
# Cierre del archivo
# ===============================================
#
file.close()
```

```python
#
# Apertura del archivo usando un Context Manager
#
with open(filename, mode="r") as text_file:
    print(text_file.read())

# no requiere .close()
```
