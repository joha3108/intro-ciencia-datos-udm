# Ingestión de Datos: Archivos CSV

En esta sección aprenderemos a trabajar con archivos CSV (Comma Separated Values), el formato más común para el intercambio de datos estructurados. Utilizaremos la librería **Pandas**, que es el estándar de la industria para manipulación de datos en Python.

```python
import pandas as pd
import numpy as np
```

## 1. Carga de datos desde archivos locales

Para cargar un CSV local, usamos la función `pd.read_csv()`. 

> **Nota:** Asegúrate de tener el archivo en la ruta correcta. Normalmente los datos crudos se guardan en una carpeta llamada `raw/` o `data/`.

```python
df = pd.read_csv('raw/datos_ejemplo.csv')
df.head()
```

## 2. Carga de datos desde URLs (Web)

Pandas permite cargar archivos directamente desde una URL. Esto es muy útil para datasets de prueba alojados en GitHub o repositorios públicos.

```python
# Dataset de ejemplo: Flores Iris
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df_iris = pd.read_csv(url)

print("Primeras 5 filas del dataset Iris:")
df_iris.head()
```

```python
df_iris.info()
```

## 3. Exploración y Filtrado Básico

Una vez cargados los datos, podemos realizar operaciones sencillas para entender su estructura.

```python
df_iris['species']
```

```python
df_iris.species
```

```python
df_iris['species'] == 'setosa'
```

```python
df_iris_setosa_petalo_largo = df_iris[(
    (df_iris['species'] == 'setosa') & 
    (df_iris['sepal_length']>=5) &
    (df_iris['petal_length']>=1)
    )]
```

```python
setosa_df = df_iris[df_iris['species'] == 'setosa']
```

```python
# Información general del DataFrame
print(df_iris.info())

# Filtrar por una condición (ej. solo flores de la especie 'setosa')
setosa_df = df_iris[df_iris['species'] == 'setosa']
print(f"\nNúmero de registros setosa: {len(setosa_df)}")

# Filtrar por valores numéricos (ej. largo de pétalo > 5)
pétalo_largo = df_iris[df_iris['petal_length'] > 5.0]
pétalo_largo.head()
```

## 4. Guardar datos procesados

Después de filtrar o limpiar nuestros datos, es común querer guardarlos en un nuevo archivo CSV.

```python
# Creamos una carpeta 'output' si no existe
import os
if not os.path.exists('output'):
    os.makedirs('output')

# Guardamos el dataframe filtrado
# index=False evita que se guarde la columna de índices automáticos de pandas
setosa_df.to_csv('output/solo_setosa.csv', index=False)
print("Archivo guardado en output/solo_setosa.csv")
```

## 5. Tips adicionales de `read_csv`

A veces los archivos no son perfectos. Aquí algunos parámetros útiles:

- `sep=';'`: Si el separador es punto y coma en vez de coma.
- `encoding='utf-8'` o `'latin-1'`: Para manejar caracteres especiales.
- `usecols=['col1', 'col2']`: Para cargar solo ciertas columnas.
- `nrows=100`: Para cargar solo las primeras 100 filas (útil con archivos gigantes).
