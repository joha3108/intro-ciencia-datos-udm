**¿Cuántas personas tienen la palabra Chief en el título de su trabajo?** Apoyese en los métodos para manipular strings y en el método `.apply()`

```python
salarios['JobTitle'].apply(lambda x: 'chief' in x.lower()).sum()
```

Este código es válido, se recomienda hacer el código mas simple y agregar typing y documentación.

```python
def contains_chief(title: str) -> bool:
    return 'chief' in title.lower()

salarios['JobTitle'].apply(contains_chief).sum()
```

**Bono: ¿Existe una correlación entre la longitud del nombre del Título del trabajo y el Salario?**

Crea una nueva columna donde se guarde la longitud del titulo del trabajo. Tip: usar el método `.apply()`

```python
salarios['JobTitle_len'] = salarios['JobTitle'].apply(len)

salarios['JobTitle_len'].head()
```

Utilice el método `.corr()` de Pandas para calcular la correlación

```python
salarios['JobTitle_len'].corr(salarios['TotalPayBenefits'])
```

Esta correlación, al ser negativa y cercana a cero, indica que la relación entre la longitud del nombre del Título y el Salario es inversa (tener un título largo no implica ganar más dinero, y en sentido contrario) y muy débil, casi inexistente. En conclusión, el salario depende de otros factores.
